## Separar o questionário dos docentes por seção/pilar


*   Ambiente de Trabalho (AT)
*   Relacionamento Interpessoal (RI)
*   Liderança (L)
*   Comunicação (C)
*   Percepção Profissional (PP)
*   Desenvolvimento Profissional (DP)
*   Recompensa (R)
*   Qualidade de vida no trabalho (QVT)





```python
#docentes
df_respostas_docentes_AT = df_resposta_docentes.iloc[:,0:6]
df_respostas_docentes_RI = df_resposta_docentes.iloc[:,6:9]
df_respostas_docentes_L = df_resposta_docentes.iloc[:,9:13]
df_respostas_docentes_C = df_resposta_docentes.iloc[:,13:21]
df_respostas_docentes_PP = df_resposta_docentes.iloc[:,21:30]
df_respostas_docentes_DP = df_resposta_docentes.iloc[:,30:33]
df_respostas_docentes_R = df_resposta_docentes.iloc[:,33:39]
df_respostas_docentes_QVT = df_resposta_docentes.iloc[:,39:49]
```

##Gerando os gráficos pie chart por seção/pilar no geral

```python
#stack() + value_counts() --> Serie object
geral_docentes_AT = df_respostas_docentes_AT.stack().value_counts().astype(float)
geral_docentes_RI = df_respostas_docentes_RI.stack().value_counts().astype(float)
geral_docentes_L = df_respostas_docentes_L.stack().value_counts().astype(float)
geral_docentes_C = df_respostas_docentes_C.stack().value_counts().astype(float)
geral_docentes_PP = df_respostas_docentes_PP.stack().value_counts().astype(float)
geral_docentes_DP = df_respostas_docentes_DP.stack().value_counts().astype(float)
geral_docentes_R = df_respostas_docentes_R.stack().value_counts().astype(float)
geral_docentes_QVT = df_respostas_docentes_QVT.stack().value_counts().astype(float)

titulos_pie_chart = ['Docentes - Ambiente de Trabalho',
                     'Docentes - Relacionamento Interpessoal',
                     'Docentes - Liderança',
                     'Docentes - Comunicação',
                     'Docentes - Perspectiva Profissional',
                     'Docentes - Desenvolvimento Profissional',
                     'Docentes - Recompensa',
                     'Docentes - Qualidade de Vida no Trabalho']
```

```python
'''
cores do pie chart

Concordo Totalmente - #3182bd
Concordo Parcialmente - #9ecae1
Indiferente - lightgray
Discordo Parcialmente - #fdbb84
Discordo Totalmente - #e34a33
'''

#plotagem AT
plot_at = geral_docentes_AT.plot.pie(autopct='%1.1f%%', 
                          ylabel='Representatividade (%)', 
                          shadow=True, 
                          title=titulos_pie_chart[0], 
                          startangle=40,
                          explode=(0.1,0,0,0,0),
                          colors=['#9ecae1','#3182bd','#fdbb84','#e34a33','lightgray'])
plot_at.figure.savefig(img_charts_path+'docentes_AT.jpg')
```

##Definindo a escala de acordo com o conteúdo do dataset

##Plotando a escala likert de forma padrão

```python
escala = ['Discordo Totalmente',
          'Discordo Parcialmente',
          'Indiferente',
          'Concordo Parcialmente',
          'Concordo Totalmente']

cores = ["#ffffff00",
         "#e34a33",
         "#fdbb84",
         "lightgray",
         "#9ecae1",
         "#3182bd"]
```

```python
#DOCENTES - AMBIENTE DE TRABALHO
#porcentagem respostas
ax_docentes_AT_1 = plot_likert.plot_likert(df_respostas_docentes_AT, escala, plot_percentage=True, colors=cores);
ax_docentes_AT_1.figure.set_size_inches(7, 6)
ax_docentes_AT_1.set_title("Docentes - Ambiente de Trabalho")
ax_docentes_AT_1.xaxis.set_label_text('Porcentagem de respostas')
ax_docentes_AT_1.figure.savefig(img_likert_scale+'docentes_AT_percentage', bbox_inches='tight')

#numero respostas
ax_docentes_AT_2 = plot_likert.plot_likert(df_respostas_docentes_AT, escala, colors=cores);
ax_docentes_AT_2.figure.set_size_inches(7, 6)
ax_docentes_AT_2.set_title("Docentes - Ambiente de Trabalho")
ax_docentes_AT_2.xaxis.set_label_text('Número de respostas')
ax_docentes_AT_2.figure.savefig(img_likert_scale+'docentes_AT_number', bbox_inches='tight')
```
